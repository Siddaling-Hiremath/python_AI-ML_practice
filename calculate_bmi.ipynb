```python
name = input("Enter name :")
age = int(input("Enter age :"))
height =input("Enter height :").strip().lower()
if "ft" in height:
    height_num=float(height.replace("ft",""))
    height_m = (height_num/3.281)
elif "cm" in height:
    height_num=float(height.replace("cm",""))
    height_m=(height_num/100)
elif "m" in height:
    height_num=float(height.replace("m",""))
    height_m=height_num
else:
    print("Enter vaild unit (ex: ft,cm,m)")
    exit()
weight =input("enter weight in kg:").strip().lower()
if "kg" in weight:
    weight= weight.replace("kg","")
    weight= float(weight)
else:
    weight= float(weight)


bmi=(weight/(height_m*height_m))
roundedbmi=round(bmi,2)
if roundedbmi < 18.5:
    status = "underweight"
elif roundedbmi>=18.5 and roundedbmi<=24.9:
    status= "normal"
elif roundedbmi>24.9 and roundedbmi<=29.9:
    status = "Overweight"
else:
    status = "Obese"
    

print("Your (BODY MASS INDEX) BMI is :",round(bmi,2))
print(f"{name} your {age} old \n height in meter {height_m} \nweight :{weight} \nyour BMI :{roundedbmi} \nstaus :{status}")
```
