```python
import json

def load_data(file_name):
    with open(file_name,"r") as f:
        data=json.load(f)
    return data
    
```

```python
data=load_data("feedback.json")
print(data)
```

```python

def clean_data(data):
    text_to_num={"one":1,"two":2,"three":3,"four":4,"five":5}
    unique_name=set()
    cleaned_data=[]
    
    for user in data:
        raw_rating= str(user["rating"]).strip().lower()
        if(raw_rating in text_to_num):
            raw_rating=text_to_num[raw_rating]
            user["rating"]=raw_rating
        #missing vlaue
        raw_age=user.get("age")
        if(raw_age == None):
            user["age"]="unknown"
        #deduplicate 
        name= user["name"].strip().lower()
        user["name"]=name
        if(name in unique_name):
            continue
        unique_name.add(name)
        cleaned_data.append(user)
    return cleaned_data
```

```python
data=clean_data(data)
```

```python
#meaningfull insights

def insights(data):
    tot_rating=0
    for user in data:
        tot_rating+=float(user["rating"])
    print(f"avg rating is {round(tot_rating/len(data),2)}")
    poor_rating=0
    for user in data:
        if (float(user["rating"])<4):
            poor_rating+=1

    print(f"percentage of user with poor rating is {round((poor_rating/len(data) *100),2)}%")
            
        
```

```python
insights(data)
```
